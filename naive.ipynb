# 4 mic setup

1. Read all 4 channels (pre recorded audio)
2. Define a windowing function
3. Compute Spectrogram

```python
from scipy.io.wavfile import read as waveread
from funcy import compose, partial
from operator import mul
from matplotlib.colors import LogNorm

SAMPLE_RATE, x1 = waveread("mdf-01.wav")
_, x2 = waveread("mdf-02.wav")
_, x3 = waveread("mdf-03.wav")
_, x4 = waveread("mdf-04.wav")

xs = [x1, x2, x3, x4]

WIN_SIZE = 2**10
HOP = int((1-3/4) * WIN_SIZE)

def windows(x, win_size=WIN_SIZE, hop=HOP):
    num_hops = int(np.floor((len(x)-win_size)/hop))
    steps = (i*hop for i in range(num_hops))
    return [x[step:win_size+step] for step in steps]

def spectrogram(x, win_size=WIN_SIZE, hop=HOP):
    to_freq = compose(np.transpose, np.matrix, np.fft.fft,
                      partial(mul, np.hamming(win_size)))

    return np.abs(np.array(np.hstack(map(to_freq, windows(x)))))[win_size/2:,:]

specs = list(map(spectrogram, xs))
```

# Compute TDOA for Each Window

```python
from numpy import concatenate
from numpy.fft import ifft, fft


def tdoa2(x0, x1):
    y = real(ifft(fft(x0)*conjugate(fft(x1))))
    right = y[0:len(y)/2]
    left = y[len(y)/2:-1]
    correlated = concatenate([left, right])
    
    center = len(correlated)/2
    max_index = argmax(correlated)
    offset = max_index - center
    
    return offset
```

```python
from itertools import combinations

tdoa_list = [[tdoa2(win1, win2) for win1, win2 in zip(a, b)]
                for a,b in combinations(map(windows, xs), r=2)]

tdoas = vstack(tdoa_list)
```

# Compute Covariance of TDOAs of the 4 channels

Treating the time delay of a channel as a random variable.
- 6 ways to combine the channels
- Results in a $6\times 6$ covariance matrix

```python
covar = cov(tdoas)
pcolor(covar)
colorbar()
```

The covariance seems to be almost completely diagonal (which is not what I expected given that the 4 mics are connected to the same system)

- The mics usually read silence, thus the background noise (which is probably not correlated) dominates the random variable

# Restricting to High Energy Windows

Windows with high energy ought to be correlated because the wave travels to all 4 mics.
If background noise is really the problem, then focusing to these windows should show much larger coorelation

```python
z = vstack([y for spec, y in zip(specs[0].T, tdoas.T) if mean(spec) > 500])
```

```python
covar = cov(z.T)
pcolor(covar)
colorbar()
show()

pcolor(abs(covar))
colorbar()
show()
```

The magnitudes of the covariances are non trivial between mics!
